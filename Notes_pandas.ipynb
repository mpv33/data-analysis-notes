# 1.1-creating a df: 
 

```python
import pandas as pd
 
# list of strings
lst = ['Geeks', 'For', 'Geeks', 'is', 
            'portal', 'for', 'Geeks']
# intialise data of lists.
data = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18]}
 
# Create DataFrame
a = pd.DataFrame(data,index=['a','b','c','d'])
b = pd.DataFrame(lst)
print(b)
print()
print(a)
```

# 1.2-Dealing with Rows and Columns:


```python
# Define a dictionary containing employee data
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
 
# Convert the dictionary into DataFrame 
df = pd.DataFrame(data)
print(df)
 
# select two columns
print(df[['Name', 'Qualification']])
print()
# select 1 row
df.loc[1]
```

# 1.3-working with missing value: <h3>Missing Data can occur when no information is provided for one or more items or for a whole unit. Missing Data is a very big problem in real life scenario. Missing Data can also refer to as NA(Not Available) values in pandas.</h3>

```python
import numpy as np
 
# dictionary of lists
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
 
# creating a dataframe from list
df = pd.DataFrame(dict)
 
# using isnull() function  
print(df.isnull())
print()
print(df.notnull())
# filling missing value using fillna()  
print(df.fillna(0))
print()
# using dropna() function  its show NAV removal from rows means fully filled value of row  
print(df)
print()
df.dropna()
```

# 1.4-Iterating over rows and columns:


```python
dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}
 
# creating a dataframe from a dictionary 
df = pd.DataFrame(dict)
 
# iterating over rows using iterrows() function 
for i, j in df.iterrows():
    print(i, j)
    print()
```

# 2.1- creating a series from list and dict:


```python
book=['math','dsa','os','cn','dbms']
num=[1,2,3,4,5]
student={'name':'mateshwari',
        'course':'b.tech',
        'branch':'cse',
        'roll_no':1648510063}
a=pd.Series(num)
b=pd.Series(book,index=['m','d','o','c','dm'])
c=pd.Series(student)
print('{}\n\n{}\n\n{}'.format(a,b,c))
```

```python
data=[1,3,5,6,7,5,7,7,1,2,4,4]
df=pd.Series(data)
print(len(df))
a=df.value_counts()
print(a)
#a.head(3)
a.tail(2)
```
